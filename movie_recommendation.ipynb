**Importing**

```python
import pandas as pd
```

```python
movies=pd.read_csv('/content/movies.csv')
```

```python
movies.columns
```

```python
movies=movies[['index','genres','id','keywords','overview','title']]
```

```python
movies
```

```python
movies['tagline'] = movies['overview']+movies['genres']
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['tagline'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
new_data[new_data['title']=="The Dark Knight Rises"]
```

```python
new_data[new_data['title']=="The Dark Knight Rises"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[3])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand('John Carter')
```
